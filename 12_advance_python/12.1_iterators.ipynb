#### Iterators are advanced python concepts that allow for efficient looping and memory management. Iterators provide a way too access elements of a collection sequentially without exposing the underlying structure.

```python
my_list = [1, 2, 3, 4, 5, 6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
iterator = iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
# iterate through all the element
next(iterator) # when all the elements were iterated by re-running it, then it gave error
```

```python
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
# string iterator
my_string = "Hello"
string_iterator = iter(my_string)

print(next(string_iterator))
print(next(string_iterator))
```
