```python
class Error(Exception):
    pass

class dobException(Error):
    pass

```

```python
year = int(input("Enter you DOB"))
age = 2026 - year

try:
    if age <= 30 and age >= 20:
        print("Valid age")

    else: 
        raise dobException
except dobException:
    print("Your age is not in the allowed bracket")

```
