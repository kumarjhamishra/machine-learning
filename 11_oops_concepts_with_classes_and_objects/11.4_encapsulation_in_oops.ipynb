```python
### Encapsulation

# with getter and setter methods
# Public, Protected, Private variables or access modifiers


class Person:
    def __init__(self, name, age):
        self.name = name # public variables
        self.age = age


def getName(person):
    return person.name

person = Person('Satyam', 20)
print(person.name)
print(person.age)
getName(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name # private variables with double under score
        self.__age = age
        self.gender = gender

def getName(person):
    return person.__name

person = Person('Krish', 20, 'Male')
#dir(person)
getName(person) # accessing private variable
# private variable can not be accessed outside the class
```

```python
# protected variabled can not be accessed outside of the class
# but can be accessed from a derived class

class Person:
    def __init__(self, name, age, gender):
        self._name = name # protected variables with single under score
        self._age = age
        self.gender = gender

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)


e = Employee('Satyam', 20, 'Male')
print(e._name)
```

```python
# Encapsulation with Getter and Setter

class Person:
    def __init__(self, name, age):
        self.__name = name # name and age are private
        self.__age = age

    # getter method for name
    # public method
    def getName(self):
        return self.__name
    
    # setter method for name
    def setName(self, name):
        self.__name = name
    
    def getAge(self):
        return self.__age   

    def setAge(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age can't be negative")
    

person = Person('Satyam', 20)

# access and modify variables using getter and setter

print(person.getName())
print(person.getAge())

person.setAge(24)
print(person.getAge())

person.setAge(-4)

```
