```python
# 1 - Create a base class named `Shape` with a method `area`. Create two derived classes `Circle` and `Square` that override the `area` 
# method. Create a list of `Shape` objects and call the `area` method on each object to demonstrate polymorphism.

import math
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return math.pi * self.r ** 2

class Square(Shape):
    def __init__(self, l):
        self.l = l

    def area(self):
        return self.l ** 2

shapes = [Circle(3), Square(4)]
for shape in shapes:
    print(shape.area())

```

```python
# 2 - Create a function named `describe_shape` that takes a `Shape` object as an argument and calls its `area` method. Create objects of 
# `Circle` and `Square` classes and pass them to the `describe_shape` function.

def describe_shape(s):
    print(f"The area of the shape is: {s.area()}")

objects = [Circle(3), Square(4)]
for obj in objects:
    describe_shape(obj)
```

```python
# 3 - Create an abstract base class named `Vehicle` with an abstract method `start_engine`. Create derived classes `Car` and `Bike` that 
# implement the `start_engine` method. Create objects of the derived classes and call the `start_engine` method.

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine has started")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine has started")

c = Car()
b = Bike()
c.start_engine()
b.start_engine()
```

```python
# 4 - In the `Vehicle` class, add a concrete method `fuel_type` that returns a generic fuel type. Override this method in `Car` and `Bike` 
# classes to return specific fuel types. Create objects of the derived classes and call the `fuel_type` method.

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    def fuel_type(self):
        print("Vehicle fuel")

class Car(Vehicle):
    def start_engine(self):
        print("Car engine has started")
    
    def fuel_type(self):
        print("Car fuel")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine has started")
    
    def fuel_type(self):
        print("Bike fuel")

c = Car()
b = Bike()
c.fuel_type()
b.fuel_type()
```
