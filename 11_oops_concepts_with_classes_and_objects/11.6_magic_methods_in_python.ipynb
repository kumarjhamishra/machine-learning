```python
# changing behaviour of __str__
class Person:
    def __init__(self,name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"{self.name}, {self.age}"

person = Person('s', 20)
print(person)
print(repr(person))
```

```python
class Person:
    def __init__(self,name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"{self.name}, {self.age}"
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person('s', 20)
print(person)
print(repr(person))
```
