```python
def isEven(num):
    """This function returns tue if the number is even and false if number is odd"""
    if num % 2 == 0:
        return True
    return False
```

```python
print(isEven(43))
```

```python
# function with multiple parameters
def add(a, b):
    """This function takes two areguments a and b and returns the sum of it"""
    return a+b
```

```python
print(add(3,2))
print(add(3,2.2))
```

```python
# default parameter
def greet(name = "Sir"):
    print(f"Hello {name}")
```

```python
greet("Suresh")
greet()
```

```python
# Variable Length Arguments

# keywords and positional arguments

# positional arguments
def printNumbers(*args):
    for num in args:
        print(num)
```

```python
printNumbers(1,2,3,4,'a', 4,5,5,"b")
```

```python
# keywords arguments -> all the parameters will be in key value pair

def printDetails(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
```

```python
printDetails(name='Satyam', age=20, sex='M', origin='India')
```

```python
def printDetails(*args, **kwargs):
    for val in args:
        print(f"Positional argument: {val}")
    
    for key, value in kwargs.items():
        print(f"{key}:{value}")
```

```python
printDetails(1,2,3,4,name='Umesh', age=40,country='India')
```
