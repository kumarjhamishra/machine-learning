small anonymous function defined using lambda keyword. Can have any number of arguments but only one expression. Commonly used for short operations or as arguments to higher order functions.

```python
addition = lambda a,b : a+b
type(addition)
```

```python
even1 = lambda num: num % 2 == 0
even1(23)
```

```python
tripleAddition = lambda a, b, c : a+b+c
tripleAddition(3,4,5)
```

```python
# map() -> function applies to all the items in a list
num = [1,2,3,4,5,6]
list(map(lambda x: x**2, num))
```
