The filter() function constructs an iterator from elements of an iterable for which a function returns true. It is used to filter out items from a list (or any other iterable) based on a condition.

```python
def checkEven(num):
    return num % 2 == 0

lst = [1,2,3,4,5,6,7,8,9]
list(filter(checkEven, lst))
```

```python
# filter with lambda function
num = [1,2,3,4,5,6,7,8,9]
greaterThanFive = list(filter(lambda x : x > 5, num))
print(greaterThanFive)
```

```python
# filter function with multiple conditions
evenAndGreaterThanFive = list(filter(lambda x : x > 5 and x % 2 == 0, num))
print(evenAndGreaterThanFive)
```

```python
# filter() to check if the age is > 25 in dictionaries
people = [
    {'name': 'Krish', 'age': 32},
    {'name': 'Jack', 'age': 23}
]

def checkAge(person):
    return person['age'] > 25

print(list(filter(checkAge, people)))
```
