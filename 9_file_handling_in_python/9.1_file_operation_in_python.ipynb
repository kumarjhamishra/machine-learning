```python
# read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line) # this space is because of the newline character after every line in the txt file
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) # .strip() function will remove the this space due to new line character
```

```python
# writing a file(Overwriting)

with open('example.txt', 'w') as file:
    file.write('Hello World\n') # new line
    file.write('this is a new line.\n') # new line
```

```python
# write a file(without overwriting)

with open('example.txt', 'a') as file:
    file.write('Append operation taking place!\n')
```

```python
# writing a list of lines to a file

lines = ['First line\n', 'Second line\n', 'Third line\n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
# Binary files - with .bin extension

# writing to a binary file in write byte mode
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
# read the content from source text file and write to a destination text file

# example.txt is already present
with open('example.txt', 'r') as source_file:
    content = source_file.read() 

# destination.txt file is not present in advance and will be created at the time of writing
with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# read a text file and count the number of lines, words and characters

# counting lines, words and characters in a text file
def countTextFile(filePath):
    with open(filePath, 'r') as file:
        lines = file.readlines()
        linesCount = len(lines)
        wordCount = sum(len(line.split()) for line in lines)
        charCount = sum(len(line) for line in lines)
    
    return linesCount, wordCount, charCount

filePath = 'example.txt'
lines, words, characters = countTextFile(filePath)
print(f"Lines: {lines}, Words: {words}, Characters: {characters}")
```

```python
# writing and then reading a file

# w+ -> open a file for both reading and writing. Id the file doesn't exist, it will be create
# If the file exists, it's content is truncated (overwritten)

with open('example.txt', 'w+') as file:
    file.write('Hello world\n')
    file.write('This is a new file\n')

    # move the file cursor to the beginning of the file
    file.seek(0)

    #read the content of file
    content = file.read()
    print(content)
```
