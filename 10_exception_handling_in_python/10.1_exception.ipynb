```python
try:
    num = int(input("Enter number"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator > 0")
except Exception as e:
    print(e)
```

```python
# if the try block is executed successfully then the else will be executed
# in the case of exception it will not

try:
    num = int(input("Enter number"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator > 0")
except Exception as e:
    print(e)

else:
    print(f"The result is {result}")
```

```python
# error aaye chaahe nhi finally to execute hoga hi

try:
    num = int(input("Enter number"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator > 0")
except Exception as e:
    print(e)

else:
    print(f"The result is {result}")

finally:
    print("Execution complete")
```

```python
### File handling and exception handling

# no matter error came or not but we need to close the file in end
try:
    file = open('example.txt', 'r')
    content = file.read()
    a = b
    print(content)
except FileNotFoundError:
    print("The file doesn't exist")
except Exception as e:
    print(e)
finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print('File closed')
```
