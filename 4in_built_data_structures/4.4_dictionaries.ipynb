```python
# unordered collection of items
# keys - unique and immutable (eg strings, numbers, tuples)
```

```python
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(type(empty_dict))
```

```python
student = {"name":"Krish", "age":32, "grade":23}
print(student)
```

```python
# with same keys - it takes the updated value
student = {"name":"Krish", "age":32, "name":23}
print(student)
```

```python
student = {"name":"Krish", "age":32, "grade":23}

# accessing values with keys
print(student['grade'])
print(student['age'])

# accesing values with get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', 'key not found'))
```

```python
# dictionary are mutable - add, update, delete

# updating value
student['age'] = 54
print(student)
# adding key value
student['address'] = 'India'
print(student)
# deleting key value pair
del student['address']
print(student)
```

```python
# Dictionary methods

# get all the keys
keys = student.keys()
print(keys)
values = student.values()
print(values)

# key value pair in items form
items = student.items()
print(items)
```

```python
# shallow copy
student_copy = student
print(student)
print(student_copy)

student['name'] = "Satyam"
print(student)
print(student_copy)
```

```python
# allocationg another memory for the copy dict
student_copy1 = student.copy() 
student['name'] = 'Shivam'
print(student)
print(student_copy1)
```

```python
# iterating over dictionaries

# iterating over keys
print("iterating over keys")
for keys in student.keys():
    print(keys)


# iterating over values
print()
print("iterating over values")
for values in student.values():
    print(values)

print()
print("Iterate over key value pairs")
# for items in student.items():
#     print(items)
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
# nested dictionaries
students = {
    "student1":{
        "name": "Satyam", "age": 20
    },
    "student2":{
        "name": "Shivam", "age": 21
    }
}

print(students)
```

```python
# accessing  nested dictionaries elements
print(students["student2"]["age"])
print(students["student2"]["name"])
```

```python
# iterating over nested dictionaries
for stu_id, stu_info in students.items():
    print(f"{stu_id}: {stu_info}")
    for key, value in stu_info.items():
        print(f"{key}: {value}")
```

```python
# dictionary comprehension
squares = {x:x*x for x in range(6)}
print(squares)
```

```python
# conditional dictionary comprehension
even_squares = {x:x*x for x in range(1, 6) if x % 2 == 0}
print(even_squares)
```

```python
# Practical examples

# use a dictionary to count the frequency of elements in list
lst = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
my_dict = {}

for num in lst:
    if num in my_dict:
        my_dict[num] += 1
    else:
        my_dict[num] = 1
print(my_dict)
```

```python
# merge 2 dictionaries into one
dict1 = {'a':1, 'b':2}
dict2 = {'b': 3, 'c': 4}
merged = {**dict1, **dict2}
print(merged)
```
