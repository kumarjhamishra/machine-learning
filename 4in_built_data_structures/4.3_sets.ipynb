```python
# membership test
st = {1,2,3,4,5}
print(3 in st)
print(15 in st)
```

```python
st1 = {1,2,3,4,5}
st2 = {4,5,6,7,9}

union = st1.union(st2)
print(union)

intersection = st1.intersection(st2)
print(intersection)

# st1.intersection_update(st2)
# print(st1)

print(st1.difference(st2))
```

```python
st1.symmetric_difference(st2)
```

```python
st1 = {1,2,3}
st2 = {4,5,6}
print(st1.issubset(st2))
print(st1.issuperset(st2))
```
