```python
# 6 - Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

import random
randomList = [random.randint(1, 40) for _ in range(30)]

randomList.sort()
print(randomList)

randomList.sort(reverse=True)
print(randomList)

# remove the duplicates
uniqueList = [randomList[0]]
for i in randomList:
    if i != uniqueList[-1]:
        uniqueList.append(i)

print(uniqueList)

```

```python
# 8 - Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. 
# Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

students = [
    {'name':'a', 'score':55},
    {'name':'b', 'score':65},
    {'name':'c', 'score':78},
    {'name':'d', 'score':45},
    {'name':'e', 'score':94},
    {'name':'f', 'score':87},
    {'name':'g', 'score':33}
]

sortedStudents = sorted(students, key=lambda x: x['score'], reverse=True)

print("Sorted students according to score in decreasing order")

for student in sortedStudents:
    print(student)
```

```python
# 9 - Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

original = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

def createTranspose(lst):
    t_lst = [[lst[j][i] for j in range(len(lst))] for i in range(len(lst[0]))]
    return t_lst
    

transpose = createTranspose(original)

print("original matrix")
for row in original:
    print(row)

print("transposed matrix")
for row in transpose:
    print(row)

    
```

```python
# 10 - Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

nestedList = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

def flattenList(nestedList):
    flatten = [item for sublist in nestedList for item in sublist]
    return flatten

singleList = flattenList(nestedList)

print("nested list")
print(nestedList)

print("single list")
print(singleList)
```

```python
# 13 - Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

originalList = [1, 2, 3, 4, 5]
def reverseList(lst):
    return lst[::-1]

reversedList = reverseList(originalList)

print(f"original list: {originalList}")
print(f"reverse list: {reversedList}")
```

```python
# 14 - Write a function that rotates a list by n positions. Print the original and rotated lists.

lst = list(range(1, 11))

def rotateList(lst, n):
    return lst[n:] + lst[:n]

print(f"original list: {lst}")
rotated = rotateList(lst, 3)
print(f"rotated list: {rotated}")
```

```python
# 15 - Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. 
# Print the intersected list.

lst1 = [1, 2, 4, 4, 6]
lst2 = [2, 3, 5]

def commonElemenets(lst1, lst2):
    return [x for x in lst1 if x in lst2]

common = commonElemenets(lst1, lst2)
print(f"common elements: {common}")
```
