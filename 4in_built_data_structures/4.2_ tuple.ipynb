```python
# tuples are immutable

# creating a tuple
tup = ()
print(tup)
print(type(tup))
```

```python
tup = tuple()
print(type(tup))
```

```python
# list to tuple
nums = tuple([1,2,3,4,5])
print(nums)
```

```python
# tuple to list
list((1,2,3,4,5))
```

```python
mixedTuple = (1, "a", 'a', 3.14, True)
mixedTuple
```

```python
# accessing tuple elements
nums = (1,2,3,4,5,6)

nums[3]
nums[-1]
```

```python
nums[0:4]
```

```python
nums[::]
```

```python
nums[::-1]
```

```python
# Tuple Operations

concatenated = nums + mixedTuple
print(concatenated)
```

```python
mixedTuple * 3
```

```python
nums * 3
```

```python
# Immutable nature of tuples

nums[1] = "umesh"
```

```python
# Tuple Methods

# count occurence
print(nums.count(1))

# index of particular element - first found from left hand side
print(nums.index(3))
```

```python
# packing and unpacking tuple

# packing
packed = 1,"Hello",3.14
print(packed)

# unpacking
a,b,c=packed
print(a)
print(b)
print(c)

# unpacking with *
numbs = (1,2,3,4,5)
first, *middle, secondLast, last = numbs
print(first)
print(middle)
print(secondLast)
print(last)
```

```python
# Nested Tuple
nestedTuple = ((1,2,3),("a","b","c"),(True, False))

# access the elements
print(nestedTuple[0])
print(nestedTuple[2][1])
```

```python
# iterating over nestes tuple
for subTuple in nestedTuple:
    for item in subTuple:
        print(item, end=" ")
    print()
```
