```python
# 1 - Create a set with the first 10 positive integers. Print the set.
st = set(range(1, 11))
print(st)
```

```python
# 2 - Add the number 11 to the set created in Assignment 1. Then remove the number 1 from the set. Print the modified set.
st.add(11)
st.remove(1)
print(st)
```

```python
# 3 - Create two sets: one with the first 5 positive integers and another with the first 5 even integers. 
# Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.

st1 = {1,2,3,4,5}
st2 = {2,4,6,8,10}

print(st1.union(st2))
print(st1.intersection(st2))
print(st1.difference(st2))
print(st1.symmetric_difference(st2))
```

```python
# 4 - Create a new set containing the squares of the first 10 positive integers using a set comprehension. Print the new set.
st = {i*i for i in range (1, 11)}
print(st)
```

```python
# 5 - Create a new set containing only the even numbers from the set created in Assignment 1 using a set comprehension. Print the new set.
st1 = {ele for ele in st if ele % 2 == 0}
print(st1)
```

```python
# 6 - Create a set with duplicate elements and remove the duplicates using set methods. Print the modified set.
s = {1, 2, 2, 3, 3, 3}
unique = set(s)
print(unique)
```

```python
# 7 - Create two sets: one with the first 5 positive integers and another with the first 3 positive integers. 
# Check if the second set is a subset of the first set and if the first set is a superset of the second set. Print the results.

st1 = set(range(1, 6))
st2 = set(range(1, 4))
print(st2.issubset(st1))
print(st1.issuperset(st2))
```

```python
# 8 - Create a frozenset with the first 5 positive integers. Print the frozenset.
fs = frozenset(range(1, 6))
print(fs)
```

```python
# 9 - Create a set with the first 5 positive integers. Convert it to a list, append the number 6, and convert it back to a set. Print the resulting set.
st = set(range(1, 6))
lst = list(st)
lst.append(6)
st = set(lst)
print(st)
```

```python
# 10 - Create a dictionary with set keys and integer values. Print the dictionary.
my_dict = {{1,2}:1, {3,4}:2, {5,6}:3}
print(my_dict)
```

```python
my_dict = {frozenset({1,2}):1, frozenset({3,4}):2, frozenset({5,6}):3}
print(my_dict)
```

```python
# 11 - Create a set and iterate over the elements, printing each element.
st = {i for i in range(1, 11)}

for ele in st:
    print(ele, end=" ")
```

```python
# 12 - Create a set and remove elements from it until it is empty. Print the set after each removal.
st = set(range(1, 11))
while len(st) > 0:
    st.pop()
    print(st)
```

```python
# 13 - Create two sets and update the first set with the symmetric difference of the two sets. Print the modified first set.
st1 = {1,2,3,4,5,6,7,8}
st2 = {4,5,6,7,8,8,8,8,8,8,9, 10, 11}

st1.symmetric_difference_update(st2)
print(st1)
```

```python
# 14 - Create a set and test if certain elements are present in the set. Print the results.
st = {1, 2, 3, 7, 8, 9}
print(f"checking for number 2: {3 in st}")
print(f"checking for number 4: {4 in st}")
```
