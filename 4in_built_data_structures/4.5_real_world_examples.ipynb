#### Create a to do list to keep track of tasks

```python
to_do_list = ['a', 'b', 'c']

# add more task
to_do_list.append('d')
to_do_list.append('e')

# remove particular task
to_do_list.remove('b')

# checking if a task is in the list
if 'a' in to_do_list:
    print("a is still to do")

print("checking remaining task")
for task in to_do_list:
    print(f"-{task}")
```

#### Create a list to store and calculate average grades for students

```python
grades = [2,3,4,5]

grades.append(34)

average_grades = sum(grades)/len(grades)
print(f"average grades: {average_grades}")

highest_grade = max(grades)
lowest_grade = min(grades)
print(f"highest grade: {highest_grade}")
print(f"lowest grade: {lowest_grade}")
```

#### Use a list to manage inventory items in a store

```python
inventory = ['apple', 'banana', 'mango']

inventory.append('water melon')

inventory.remove('banana')

# checking if an item is in stock
item = 'mango'

if item in inventory:
    print(f"{item} are in stock")
else:
    print(f"{item} not in stock")

# printing the inventory
print("inventory list")
for item in inventory:
    print(f"- {item}")
```

#### Use a list to collect and analyze user feedback

```python
feedback = ["good", "best", "fine", "badhiya"]

feedback.append("not happy")

# counting specific feedback
positive_feedback_count = sum(1 for comment in feedback if "good" in comment.lower() or "best" in comment.lower() or "fine" in comment.lower() or "badhiya" in comment.lower())
print(f"positive feedback count: {positive_feedback_count}")

print("user feedback")
for comment in feedback:
    print(f"- {comment}")
```
